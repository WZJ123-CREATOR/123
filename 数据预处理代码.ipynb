```python
# 导入必要的库
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# 设置中文显示
plt.rcParams['font.sans-serif'] = ['SimHei']  # 用来正常显示中文标签
plt.rcParams['axes.unicode_minus'] = False  # 用来正常显示负号

# 读取数据（假设数据已经保存为Excel文件）
try:
    # 尝试从当前目录读取
    df = pd.read_excel("广州海底捞真实评论数据.xlsx")
except FileNotFoundError:
    # 如果当前目录没有，尝试从桌面读取
    desktop_path = os.path.join(os.path.expanduser("~"), "Desktop", "广州海底捞真实评论数据.xlsx")
    df = pd.read_excel(desktop_path)

# ================== 评分一致性检查 ===========# ...省略其他代码... 


# ==== 修改后代码 ====

# 1. 检查评分范围是否合理（1-5分）
print("===== 评分范围检查 =====")
invalid_ratings = df[(df['评分'] < 1) | (df['评分'] > 5)]
print(f"无效评分数量: {len(invalid_ratings)} 条")
if len(invalid_ratings) > 0:
    print("无效评分示例:")
    print(invalid_ratings[['评论ID', '评分']].head())

# 2. 检查评分分布是否符合预期（正评60%，中评30%，负评10%）
print("\n===== 评分分布检查 =====")
rating_distribution = df['评分'].value_counts().sort_index()
print("实际评分分布:")
print(rating_distribution)

# 计算预期评分分布
total_reviews = len(df)
expected_positive = int(total_reviews * 0.6)
expected_neutral = int(total_reviews * 0.3)
expected_negative = total_reviews - expected_positive - expected_neutral

print("\n预期评分分布:")
print(f"正面评价: {expected_positive} 条")
print(f"中性评价: {expected_neutral} 条")
print(f"负面评价: {expected_negative} 条")

# 可视化评分分布
plt.figure(figsize=(10, 6))
sns.histplot(df['评分'], bins=10, kde=True)
plt.title('评论评分分布')
plt.xlabel('评分')
plt.ylabel('数量')
plt.xticks(np.arange(1, 6, 0.5))
plt.tight_layout()
plt.show()

# 3. 检查评分与评论内容的情感一致性
print("\n===== 评分与评论内容情感一致性检查 =====")
# 简单的情感分析（基于关键词）
positive_keywords = ['太赞了', '必须给五星好评', '绝了', '惊艳', '满意', '好吃', '新鲜', '正宗', '舒适', '热情']
negative_keywords = ['失望', '糟糕', '差', '嘈杂', '不好', '不满意', '不新鲜', '不正宗', '不舒适', '冷淡']

df['情感倾向'] = df['评论内容'].apply(
    lambda x: 'positive' if any(keyword in x for keyword in positive_keywords) 
              else ('negative' if any(keyword in x for keyword in negative_keywords) 
                    else 'neutral')
)

# 检查评分与情感倾向是否一致
inconsistent_ratings = df[
    ((df['评分'] >= 4.0) & (df['情感倾向'] == 'negative')) | 
    ((df['评分'] <= 2.0) & (df['情感倾向'] == 'positive'))
]
print(f"评分与情感倾向不一致的数量: {len(inconsistent_ratings)} 条")
if len(inconsistent_ratings) > 0:
    print("评分与情感倾向不一致的示例:")
    print(inconsistent_ratings[['评论ID', '评分', '情感倾向', '评论内容']].head())

# 可视化评分与情感倾向的关系
plt.figure(figsize=(10, 6))
sns.countplot(x='评分', hue='情感倾向', data=df)
plt.title('评分与情感倾向关系')
plt.xlabel('评分')
plt.ylabel('数量')
plt.xticks(np.arange(1, 6))
plt.legend(title='情感倾向')
plt.tight_layout()
plt.show()

# 4. 检查不同平台的评分分布是否一致
print("\n===== 不同平台评分分布检查 =====")
platform_rating = df.groupby('平台')['评分'].mean()
print("各平台平均评分:")
print(platform_rating)

# 可视化不同平台评分分布
plt.figure(figsize=(8, 6))
platform_rating.plot(kind='bar')
plt.title('各平台平均评分')
plt.xlabel('平台')
plt.ylabel('平均评分')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()

# 5. 检查不同餐厅的评分分布是否一致
print("\n===== 不同餐厅评分分布检查 =====")
restaurant_rating = df.groupby('餐厅')['评分'].mean().sort_values(ascending=False)
print("各餐厅平均评分（前10）:")
print(restaurant_rating.head(10))

# 可视化不同餐厅评分分布
plt.figure(figsize=(12, 8))
restaurant_rating.head(10).plot(kind='bar')
plt.title('各餐厅平均评分（前10）')
plt.xlabel('餐厅')
plt.ylabel('平均评分')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# 6. 检查不同菜品的评分分布是否一致
print("\n===== 不同菜品评分分布检查 =====")
dish_rating = df.groupby('菜品')['评分'].mean().sort_values(ascending=False)
print("各菜品平均评分（前10）:")
print(dish_rating.head(10))

# 可视化不同菜品评分分布
plt.figure(figsize=(12, 8))
dish_rating.head(10).plot(kind='bar')
plt.title('各菜品平均评分（前10）')
plt.xlabel('菜品')
plt.ylabel('平均评分')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# 总结报告
print("\n===== 评分一致性验证总结 =====")
print(f"无效评分数量: {len(invalid_ratings)} 条")
print(f"评分与情感倾向不一致的数量: {len(inconsistent_ratings)} 条")

if len(invalid_ratings) > 0 or len(inconsistent_ratings) > 0:
    print("\n发现的问题:")
    if len(invalid_ratings) > 0:
        print(f"- 存在{len(invalid_ratings)}条无效评分（不在1-5分范围内）")
    if len(inconsistent_ratings) > 0:
        print(f"- 存在{len(inconsistent_ratings)}条评分与情感倾向不一致的评论")
else:
    print("\n所有评分均符合一致性要求！")

print("\n建议:")
if len(invalid_ratings) > 0:
    print("- 请检查数据生成代码，确保所有评分都在1-5分范围内")
if len(inconsistent_ratings) > 0:
    print("- 请检查评分生成逻辑或评论模板，确保评分与评论内容情感一致")
```

```python
# 导入必要的库
import pandas as pd
import numpy as np
import os

# 读取数据（假设数据已经保存为Excel文件）
try:
    # 尝试从当前目录读取
    df = pd.read_excel("广州海底捞真实评论数据.xlsx")
except FileNotFoundError:
    # 如果当前目录没有，尝试从桌面读取
    desktop_path = os.path.join(os.path.expanduser("~"), "Desktop", "广州海底捞真实评论数据.xlsx")
    df = pd.read_excel(desktop_path)

# ================== 数据清洗 ===========# ...省略其他代码... 


# ==== 修改后代码 ====

# 1. 缺失值处理
print("===== 缺失值处理 =====")
print("原始数据缺失值情况：")
print(df.isnull().sum())

# 检查是否有缺失值
if df.isnull().any().any():
    # 对于数值型列（如评分），用中位数填充
    numeric_cols = df.select_dtypes(include=[np.number]).columns
    for col in numeric_cols:
        median_value = df[col].median()
        df[col].fillna(median_value, inplace=True)
        print(f"列 '{col}' 的缺失值已用中位数 {median_value} 填充")
    
    # 对于文本型列（如评论内容），用"无评论内容"填充
    text_cols = df.select_dtypes(exclude=[np.number]).columns
    for col in text_cols:
        if col != '评论ID' and col != '平台' and col != '餐厅' and col != '菜品' and col != '日期':
            df[col].fillna("无评论内容", inplace=True)
            print(f"列 '{col}' 的缺失值已用'无评论内容'填充")
else:
    print("数据中没有缺失值")

# 2. 异常值修正
print("\n===== 异常值修正 =====")
# 检查评分是否在合理范围内（1-5分）
invalid_ratings = df[(df['评分'] < 1) | (df['评分'] > 5)]
if not invalid_ratings.empty:
    print(f"发现 {len(invalid_ratings)} 条无效评分（不在1-5分范围内）")
    
    # 修正异常评分：将小于1的设为1，大于5的设为5
    df['评分'] = df['评分'].apply(lambda x: 1 if x < 1 else (5 if x > 5 else x))
    print("异常评分已修正为1-5分范围内")
else:
    print("没有发现异常评分")

# 3. 去重处理
print("\n===== 去重处理 =====")
# 检查完全重复的行
duplicates = df[df.duplicated(keep=False)]
if not duplicates.empty:
    print(f"发现 {len(duplicates)} 条完全重复的行")
    
    # 删除所有完全重复的行，只保留一条
    df = df.drop_duplicates(keep='first')
    print("已删除所有完全重复的行")
else:
    print("没有发现完全重复的行")

# 检查评论ID是否唯一（理论上应该唯一）
duplicate_ids = df[df.duplicated(subset=['评论ID'], keep=False)]
if not duplicate_ids.empty:
    print(f"发现 {len(duplicate_ids)} 条评论ID重复的记录")
    
    # 保留第一条记录，删除其他重复的评论ID
    df = df.drop_duplicates(subset=['评论ID'], keep='first')
    print("已删除重复的评论ID记录")
else:
    print("评论ID都是唯一的")

# 检查平台+餐厅+菜品组合是否有重复（理论上不应该重复）
duplicate_combinations = df[df.duplicated(subset=['平台', '餐厅', '菜品'], keep=False)]
if not duplicate_combinations.empty:
    print(f"发现 {len(duplicate_combinations)} 条平台+餐厅+菜品组合重复的记录")
    
    # 保留第一条记录，删除其他重复的组合
    df = df.drop_duplicates(subset=['平台', '餐厅', '菜品'], keep='first')
    print("已删除重复的平台+餐厅+菜品组合记录")
else:
    print("平台+餐厅+菜品组合都是唯一的")

# ================== 数据清洗后检查 ===========# ...省略其他代码... 


# ==== 修改后代码 ====

# 检查清洗后的数据情况
print("\n===== 数据清洗后检查 =====")
print(f"清洗后数据行数: {df.shape[0]}")
print(f"清洗后数据列数: {df.shape[1]}")

# 再次检查缺失值
print("\n清洗后缺失值情况：")
print(df.isnull().sum())

# 检查评分范围
print("\n清洗后评分范围检查：")
print(df['评分'].describe())

# 检查评论ID唯一性
print("\n清洗后评论ID唯一性检查：")
print(df['评论ID'].duplicated().sum())

# 检查平台+餐厅+菜品组合唯一性
print("\n清洗后平台+餐厅+菜品组合唯一性检查：")
print(df.duplicated(subset=['平台', '餐厅', '菜品']).sum())

# ================== 保存清洗后的数据 =# ...省略其他代码... 


# ==== 修改后代码 ====
try:
    cleaned_filename = "广州海底捞真实评论数据_清洗后.xlsx"
    df.to_excel(cleaned_filename, index=False)
    print(f"\n清洗后的数据已保存到: {cleaned_filename}")
except PermissionError:
    desktop_path = os.path.join(os.path.expanduser("~"), "Desktop", "广州海底捞真实评论数据_清洗后.xlsx")
    df.to_excel(desktop_path, index=False)
    print(f"\n清洗后的数据已保存到桌面: {desktop_path}")

# 显示清洗后的前5条数据
print("\n===== 清洗后的数据示例 =====")
print(df.head())
```

```python
# 导入必要的库
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# 设置中文显示
plt.rcParams['font.sans-serif'] = ['SimHei']  # 用来正常显示中文标签
plt.rcParams['axes.unicode_minus'] = False  # 用来正常显示负号

# 读取数据（假设数据已经保存为Excel文件）
try:
    # 尝试从当前目录读取
    df = pd.read_excel("广州海底捞真实评论数据.xlsx")
except FileNotFoundError:
    # 如果当前目录没有，尝试从桌面读取
    desktop_path = os.path.join(os.path.expanduser("~"), "Desktop", "广州海底捞真实评论数据.xlsx")
    df = pd.read_excel(desktop_path)

# ================== 初步统计分析 ===========# ...省略其他代码... 


# ==== 修改后代码 ====

# 1. 数据概览
print("===== 数据概览 =====")
print(f"数据总行数: {df.shape[0]}")
print(f"数据列数: {df.shape[1]}")
print("\n各列数据类型和缺失值情况:")
print(df.info())

# 2. 评分分析
print("\n===== 评分分析 =====")
# 评分分布统计
rating_stats = df['评分'].describe()
print("评分统计描述:")
print(rating_stats)

# 评分分布直方图
plt.figure(figsize=(10, 6))
sns.histplot(df['评分'], bins=10, kde=True)
plt.title('评论评分分布')
plt.xlabel('评分')
plt.ylabel('数量')
plt.xticks(np.arange(1, 6, 0.5))
plt.tight_layout()
plt.show()

# 评分分布饼图（按评分区间）
rating_bins = [1, 2, 3, 4, 5]
rating_labels = ['1-2分', '2-3分', '3-4分', '4-5分']
df['评分区间'] = pd.cut(df['评分'], bins=rating_bins, labels=rating_labels)
rating_distribution = df['评分区间'].value_counts().sort_index()
print("\n评分区间分布:")
print(rating_distribution)

plt.figure(figsize=(8, 6))
rating_distribution.plot(kind='pie', autopct='%1.1f%%', startangle=90)
plt.title('评分区间分布')
plt.ylabel('')
plt.tight_layout()
plt.show()

# 3. 平台分析
print("\n===== 平台分析 =====")
# 各平台评论数量
platform_counts = df['平台'].value_counts()
print("各平台评论数量:")
print(platform_counts)

# 平台评论占比
platform_percentage = df['平台'].value_counts(normalize=True) * 100
print("\n各平台评论占比:")
print(platform_percentage)

# 平台评论数量柱状图
plt.figure(figsize=(8, 6))
platform_counts.plot(kind='bar')
plt.title('各平台评论数量')
plt.xlabel('平台')
plt.ylabel('评论数量')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()

# 4. 餐厅分析
print("\n===== 餐厅分析 =====")
# 各餐厅评论数量
restaurant_counts = df['餐厅'].value_counts().head(10)  # 只显示前10
print("各餐厅评论数量（前10）:")
print(restaurant_counts)

# 餐厅评论占比
restaurant_percentage = df['餐厅'].value_counts(normalize=True) * 100
print("\n各餐厅评论占比（前10）:")
print(restaurant_percentage.head(10))

# 餐厅评论数量柱状图
plt.figure(figsize=(12, 8))
restaurant_counts.plot(kind='bar')
plt.title('各餐厅评论数量（前10）')
plt.xlabel('餐厅')
plt.ylabel('评论数量')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# 5. 菜品分析
print("\n===== 菜品分析 =====")
# 各菜品评论数量
dish_counts = df['菜品'].value_counts().head(10)  # 只显示前10
print("各菜品评论数量（前10）:")
print(dish_counts)

# 菜品评论占比
dish_percentage = df['菜品'].value_counts(normalize=True) * 100
print("\n各菜品评论占比（前10）:")
print(dish_percentage.head(10))

# 菜品评论数量柱状图
plt.figure(figsize=(12, 8))
dish_counts.plot(kind='bar')
plt.title('各菜品评论数量（前10）')
plt.xlabel('菜品')
plt.ylabel('评论数量')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# 6. 评论类型分析（基于评分）
print("\n===== 评论类型分析 =====")
# 根据评分划分评论类型
def get_comment_type(score):
    if score >= 4.0:
        return '正面评价'
    elif score <= 2.0:
        return '负面评价'
    else:
        return '中性评价'

df['评论类型'] = df['评分'].apply(get_comment_type)

# 各评论类型数量
comment_type_counts = df['评论类型'].value_counts()
print("各评论类型数量:")
print(comment_type_counts)

# 评论类型占比
comment_type_percentage = df['评论类型'].value_counts(normalize=True) * 100
print("\n各评论类型占比:")
print(comment_type_percentage)

# 评论类型饼图
plt.figure(figsize=(8, 6))
comment_type_counts.plot(kind='pie', autopct='%1.1f%%', startangle=90)
plt.title('评论类型分布')
plt.ylabel('')
plt.tight_layout()
plt.show()

# 7. 日期分析
print("\n===== 日期分析 =====")
# 生成日期范围
start_date = datetime.now() - timedelta(days=365)
end_date = datetime.now()

# 按月统计评论数量
df['日期'] = pd.to_datetime(df['日期'])
monthly_reviews = df.set_index('日期').resample('M').size()
print("每月评论数量统计:")
print(monthly_reviews)

# 月度评论数量折线图
plt.figure(figsize=(12, 6))
monthly_reviews.plot(kind='line', marker='o')
plt.title('每月评论数量')
plt.xlabel('日期')
plt.ylabel('评论数量')
plt.grid(True)
plt.tight_layout()
plt.show()

# 8. 综合分析
print("\n===== 综合分析 =====")
# 计算各平台在各餐厅的评论数量
platform_restaurant = df.groupby(['平台', '餐厅']).size().unstack()
print("各平台在各餐厅的评论数量:")
print(platform_restaurant)

# 计算各平台对各菜品的评论数量
platform_dish = df.groupby(['平台', '菜品']).size().unstack()
print("\n各平台对各菜品的评论数量:")
print(platform_dish)

# 生成分析报告
print("\n===== 分析报告 =====")
print(f"1. 数据总览: 共{df.shape[0]}条评论，包含{df.shape[1]}个字段")
print(f"2. 评分分布: 平均评分{rating_stats['mean']:.1f}，中位数{rating_stats['50%']:.1f}")
print(f"3. 平台分布: 美团评论{platform_counts.get('美团', 0)}条，大众点评评论{platform_counts.get('大众点评', 0)}条")
print(f"4. 餐厅分布: 最热门餐厅'{restaurant_counts.index[0]}'有{restaurant_counts.iloc[0]}条评论")
print(f"5. 菜品分布: 最热门菜品'{dish_counts.index[0]}'有{dish_counts.iloc[0]}条评论")
print(f"6. 评论类型: 正面评价{comment_type_counts.get('正面评价', 0)}条，负面评价{comment_type_counts.get('负面评价', 0)}条")

# 保存分析结果到Excel
try:
    analysis_filename = "广州海底捞评论数据分析结果.xlsx"
    with pd.ExcelWriter(analysis_filename) as writer:
        df.to_excel(writer, sheet_name='原始数据', index=False)
        rating_stats.to_frame().to_excel(writer, sheet_name='评分统计')
        platform_counts.to_frame().to_excel(writer, sheet_name='平台统计')
        restaurant_counts.to_frame().to_excel(writer, sheet_name='餐厅统计')
        dish_counts.to_frame().to_excel(writer, sheet_name='菜品统计')
        comment_type_counts.to_frame().to_excel(writer, sheet_name='评论类型统计')
    print(f"\n分析结果已保存到: {analysis_filename}")
except PermissionError:
    desktop_path = os.path.join(os.path.expanduser("~"), "Desktop", "广州海底捞评论数据分析结果.xlsx")
    with pd.ExcelWriter(desktop_path) as writer:
        df.to_excel(writer, sheet_name='原始数据', index=False)
        rating_stats.to_frame().to_excel(writer, sheet_name='评分统计')
        platform_counts.to_frame().to_excel(writer, sheet_name='平台统计')
        restaurant_counts.to_frame().to_excel(writer, sheet_name='餐厅统计')
        dish_counts.to_frame().to_excel(writer, sheet_name='菜品统计')
        comment_type_counts.to_frame().to_excel(writer, sheet_name='评论类型统计')
    print(f"\n分析结果已保存到桌面: {desktop_path}")
```
